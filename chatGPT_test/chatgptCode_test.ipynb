```python
# Import necessary libraries
from undetected_chromedriver import Chrome, ChromeOptions
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import pandas as pd

# Set up Chrome with undetected-chromedriver
options = ChromeOptions()
options.add_argument("--disable-blink-features=AutomationControlled")
driver = Chrome(options=options)

# Navigate to YouTube and click the sign-in button
driver.get("https://www.youtube.com/")
sign_in_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="buttons"]/ytd-button-renderer/yt-button-shape/a/yt-touch-feedback-shape/div/div[2]')))
sign_in_button.click()

Enter email and click next button
email_input = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "identifierId")))
email_input.send_keys("kjaehw0207")
next_button = driver.find_element(By.ID, "identifierNext")
next_button.click()

Enter password and click next button
password_input = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.NAME, "password")))
password_input.send_keys("flapaqj1!%")
next_button = driver.find_element_by_id("passwordNext")
next_button.click()

# Handle 2nd authorization
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "headingText")))
skip_button = driver.find_element_by_xpath("//button[@data-dismiss='alert']")
skip_button.click()

# Input for data collection start point
start_date = "2022-01-01" # example start date

# Navigate to watch history page
driver.get("https://www.youtube.com/feed/history")

# Collect data from watch history
video_titles = []
channels = []
running_times = []
watch_dates = []
watch_times = []

while True:
    # Get video containers on page
    video_containers = driver.find_elements_by_xpath("//ytd-video-renderer")

    # Collect data from each video container
    for video_container in video_containers:
        video_title = video_container.find_element_by_id("video-title").text
        channel = video_container.find_element_by_xpath(".//ytd-channel-name//a").text
        running_time = video_container.find_element_by_xpath(".//ytd-thumbnail-overlay-time-status-renderer").text
        watch_date = video_container.find_element_by_xpath(".//ytd-video-meta-block//span[1]").text
        watch_time = video_container.find_element_by_xpath(".//ytd-video-meta-block//span[2]").text
        
        # Check if video was watched after input start date
        if watch_date >= start_date:
            video_titles.append(video_title)
            channels.append(channel)
            running_times.append(running_time)
            watch_dates.append(watch_date)
            watch_times.append(watch_time)
        else:
            # Stop data collection if video was watched before input start date
            break

    # Check if there are more videos on the page to load
    try:
        more_button = driver.find_element_by_xpath("//yt-next-continuation//button")
        more_button.click()
    except:
        # End of watch history
        break

# Create dataframe with collected data
data = {
    "Video Title": video_titles,
    "Channel": channels,
    "Running Time": running_times,
    "Watch Date": watch_dates,
    "Watch Time": watch_times
}
df = pd.DataFrame(data)

# Save dataframe to csv
df.to
```
